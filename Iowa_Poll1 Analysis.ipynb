```python
# This line is used to redact the actual numbers used.  All numbers began with 515200

f = open('Iowa_Poll1.csv')
redacted_f = []
nums = {}
ticker = 0
responses = []
for line in f.readlines():
    line = line.split(',')
    if(line[0] == '0'):
        responses.append(line)
    if(line[0] == '0' or line[0] == '1'):
        if(line[1] not in nums.keys()):
            nums[line[1]] = ticker
            line[1] = ticker
            ticker += 1
        else:
            line[1] = nums[line[1]]
    redacted_f.append(line)
f = open('Iowa_Poll1_Redacted.csv', 'w')

for line in redacted_f:
    f.write((str)(line))
f.close()
```

```python
# Here we looked at all human responses
import matplotlib.pyplot as plt
for l in responses:
    if(len(l[3]) < 20):
        print(l[3])
```
